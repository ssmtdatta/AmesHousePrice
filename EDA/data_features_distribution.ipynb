### Notebook Summary
* Inspect how features are distributed

### Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

### Directory

```python
BASE_DIR = "../AmesHousing/"
DATA_IN = BASE_DIR+"DataDwn/"
```

### Training Data

```python
trn = pd.read_csv(DATA_IN+"train.csv")
trn.shape
```

### Exploratory

### 1 Target variable: Sales Price

```python
sns.distplot(trn['SalePrice'])
plt.xlabel("Sale Price (linear scale)")
plt.ylabel("Count")
plt.show()
```

```python
sns.distplot(trn['SalePrice'].apply(np.log))
plt.xlabel("Sale Price (log scale)")
plt.ylabel("Count")
plt.show()
```

**Note:** Log-transform gives more near normal distribution.

### 2 Feature variables
Compare the correlation between sq-ft area of individual features and total area. E.g. 'LotArea','LotFrontage' vs. total area.

#### 2A Lot area

```python
y = trn['SalePrice']

fig = plt.figure(figsize=(12, 8))

ax1 = fig.add_subplot(221)
ax1.scatter(trn['LotArea'], y)
ax1.set_ylabel("Sale Price")
ax1.set_xlabel("Lot Area")

ax2 = fig.add_subplot(222)
ax2.scatter(trn['LotFrontage'], y)
ax2.set_ylabel("Sale Price")
ax2.set_xlabel("Lot Frontage")

ax3 = fig.add_subplot(223)
ax3.scatter(trn['LotArea'], np.log(y))
ax3.set_ylabel("Log Sale Price")
ax3.set_xlabel("Lot Area")

ax4 = fig.add_subplot(224)
ax4.scatter(trn['LotFrontage'], np.log(y))
ax4.set_ylabel("Log Sale Price")
ax4.set_xlabel("Lot Frontage")

plt.tight_layout()
plt.show()
```

#### 2B Floor Area

```python
y = trn['SalePrice']

fig = plt.figure(figsize=(12, 8))

ax1 = fig.add_subplot(221)
ax1.scatter(trn['1stFlrSF'], y)
ax1.set_ylabel("Sale Price")
ax1.set_xlabel("1st Floor Area")

ax2 = fig.add_subplot(222)
ax2.scatter(trn['2ndFlrSF'], y)
ax2.set_ylabel("Sale Price")
ax2.set_xlabel("2nd Floor Area")

ax3 = fig.add_subplot(223)
ax3.scatter(trn['LowQualFinSF'], y)
ax3.set_ylabel("Sale Price")
ax3.set_xlabel("Low Quality Area")

ax4 = fig.add_subplot(224)
ax4.scatter(trn['GrLivArea'], y)
ax4.set_ylabel("Sale Price")
ax4.set_xlabel("Living Area")

plt.tight_layout()
plt.show()
```

#### 2C Overall Quality (categorical)

```python
# variation in sales price based on overall quality
y = trn['SalePrice']
x = trn['OverallQual']

f, ax = plt.subplots(figsize=(8, 6))
fig = sns.boxplot(x=x, y=y)
fig.axis(ymin=0, ymax=800000);
```

#### 2D Overall Condition (categorical)

```python
# variation in sales price based on overall condition
y = trn['SalePrice']
x = trn['OverallCond']

f, ax = plt.subplots(figsize=(8, 6))
fig = sns.boxplot(x=x, y=y)
fig.axis(ymin=0, ymax=800000);
```

#### 2E Year remodelled/built

```python
# newer or renovated houses seem to have a higher price
y = trn['SalePrice']
x = trn['YearRemodAdd']

f, ax = plt.subplots(figsize=(18, 6))
fig = sns.boxplot(x=x, y=y)
fig.axis(ymin=0, ymax=800000);
```

### Correlation between numeric features and sale price

```python
#scatterplot showing correlation between sale price and selected numerical features
num_df = pd.DataFrame()
df = trn
df = df.fillna(0)

feature_lot = ['LotArea','LotFrontage']
feature_floor = ['1stFlrSF', '2ndFlrSF','LowQualFinSF', 'GrLivArea']
feature_basement = ['BsmtFinSF1','BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF']
feature_bathroom = ['BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath']
feature_garage = ['GarageArea', 'GarageCars']
feature_porch = ['WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch', '3SsnPorch', 'ScreenPorch']


num_df['TotLotSF'] = df['LotArea'] + df['LotFrontage']
num_df['TotFlrSF'] = df['1stFlrSF']+ df['2ndFlrSF'] + df['GrLivArea'] + df['LowQualFinSF']
num_df['TotalBsmtSF'] = df['TotalBsmtSF']
num_df['TotNumBath'] = df['BsmtFullBath']+df['BsmtHalfBath']+df['FullBath']+df['HalfBath']
num_df['GarageArea'] = df['GarageArea']
num_df['PorchArea'] = df['WoodDeckSF']+df['OpenPorchSF']+df['EnclosedPorch']+df['3SsnPorch']+df['ScreenPorch']
num_df['PoolArea'] = df['PoolArea']
num_df['SalePrice'] = df['SalePrice']


num_df.shape
```
